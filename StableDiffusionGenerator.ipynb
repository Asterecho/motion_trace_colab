## ②-A　通知音のダウンロード

通知音を[効果音ラボ](https://soundeffect-lab.info/)様よりダウンロードして、セルの実行を開始・完了・失敗した時に音が鳴るようにします。

下の【②-A】のセルを実行してください



```python
#@markdown ■■■■■■■■■■■■■■■■■■

#@markdown 【②-A】　通知音のダウンロード

#@markdown ■■■■■■■■■■■■■■■■■■

from enum import Enum

class SoundType(Enum):
    START = 0
    SUCCESS = 1
    FAIL = 2

# ----------------------------------

! wget --no-check-certificate -c "https://soundeffect-lab.info/sound/anime/mp3/sceneswitch1.mp3"
! wget --no-check-certificate -c "https://soundeffect-lab.info/sound/anime/mp3/incorrect1.mp3"
! wget --no-check-certificate -c "https://soundeffect-lab.info/sound/anime/mp3/switch1.mp3"

from IPython.display import Audio, display
import torch

def play_sound(sound_type: SoundType, autoplay=True):
    try:
        if sound_type == SoundType.START:
            file_name = "/content/switch1.mp3"
        elif sound_type == SoundType.SUCCESS:
            file_name = "/content/sceneswitch1.mp3"
        else:
            file_name = "/content/incorrect1.mp3"

        display(Audio(file_name, autoplay=autoplay, normalize=False))
    except:
        print("■■■■■■■■■■■■■■■")
        print("■　効果音が再生できませんでした")
        print("■■■■■■■■■■■■■■■")

# ----------------------------------

exec_dict = {
    '②-A': {'exec': False, 'premise': []},
    '②-B': {'exec': False, 'premise': ['②-A', ]},
    '②-C': {'exec': False, 'premise': ['②-A', '②-B', ]},
    '②-D': {'exec': False, 'premise': ['②-A', '②-B', '②-C', ]},
    '②-E': {'exec': False, 'premise': ['②-A', '②-B', '②-C', '②-D', ]},
    '③-A': {'exec': False, 'premise': ['②-A', '②-B', '②-C', '②-D', '②-E', ]},
    '③-B': {'exec': False, 'premise': ['②-A', '②-B', '②-C', '②-D', '②-E', '③-A', ]},
    '④-A': {'exec': False, 'premise': ['②-A', '②-B', '②-C', '②-D', '②-E', ]},
    '④-B': {'exec': False, 'premise': ['②-A', '②-B', '②-C', '②-D', '②-E', '④-A', ]},
    '⑤-A': {'exec': False, 'premise': ['②-A', '②-B', '②-C', '②-D', '②-E', ]},
    '⑤-B': {'exec': False, 'premise': ['②-A', '②-B', '②-C', '②-D', '②-E', '⑤-A', ]},
    '⑥-A': {'exec': False, 'premise': ['②-A', '②-B', '②-C', '②-D', '②-E', ]},
    '⑥-B': {'exec': False, 'premise': ['②-A', '②-B', '②-C', '②-D', '②-E', '⑥-A', ]},
}

NVIDIA_VISIBLE_DEVICES=all

class IpyExit(SystemExit):
    def __init__(self):
        play_sound(SoundType.FAIL)
        pass

def check_exec_dict(cell_key: str):
    if not torch.cuda.is_available():
        print("■■■■■■■■■■■■■■■")
        print("■　** ERROR **")
        print("■　GPUが有効になってません。")
        print("■　準備編を参考にしながらランタイムをGPUに変更してください。")
        print("■■■■■■■■■■■■■■■")
        raise IpyExit

    if not exec_dict:
        print("■■■■■■■■■■■■■■■")
        print("■　** ERROR **")
        print("■　セル【②-A】が実行されていない可能性があります。")
        print("■　目次から戻って実行してください。")
        print("■■■■■■■■■■■■■■■")
        raise IpyExit

    for p in exec_dict[cell_key]['premise']:
        # 事前に実行されるべきセルがすべて実行されているか
        if not exec_dict[p]['exec']:
            print("■■■■■■■■■■■■■■■")
            print("■　** ERROR **")
            print(f"■　セル【{p}】が実行されていない可能性があります。")
            print("■　目次から戻って実行してください。")
            print("■■■■■■■■■■■■■■■")
            raise IpyExit

    # 全部実行されていたらOK
    play_sound(SoundType.START)
    return True

def finish_cell(cell_key: str):
    play_sound(SoundType.SUCCESS)

    print("■■■■■■■■■■■■■■■")
    print("■　** OK **")
    print(f"■　セル【{cell_key}】の実行に成功しました。")
    print("■■■■■■■■■■■■■■■")

    exec_dict[cell_key]['exec'] = True

# ----------------------------------

# セル実行可否
ckey = "②-A"
check_exec_dict(ckey)

# ----------------------------------

# セル終了

finish_cell(ckey)

print("")
print("")
print("■ セル実行開始")
play_sound(SoundType.START, autoplay=False)

print("")
print("")
print("■ セル実行失敗")
play_sound(SoundType.FAIL, autoplay=False)

print("")
print("")
print("■ セル実行成功")
play_sound(SoundType.SUCCESS, autoplay=False)
```

```python
#@markdown ■■■■■■■■■■■■■■■■■■

#@markdown 【②-E】　Tokenの認証と学習モデルのダウンロード

#@markdown ■■■■■■■■■■■■■■■■■■

#@markdown 学習モデル
difussion_model = "CompVis/stable-diffusion-v1-4"  #@param ['CompVis/stable-diffusion-v1-4', 'hakurei/waifu-diffusion']

#@markdown  - [CompVis/stable-diffusion-v1-4](https://huggingface.co/CompVis/stable-diffusion-v1-4): デフォルトの Stable Diffusion モデル

#@markdown  - [hakurei/waifu-diffusion](https://huggingface.co/hakurei/waifu-diffusion) : 「[Danbooru](https://danbooru.donmai.us)」チューニングアニメ・マンガ特化モデル

# - [naclbit/trinart_stable_diffusion](https://huggingface.co/naclbit/trinart_stable_diffusion) : 「[とりんさまAI](https://twitter.com/trinsama)」チューニングアニメ・マンガ特化モデル

trinart = difussion_model == "naclbit/trinart_stable_diffusion"

# セル実行可否
ckey = "②-E"
check_exec_dict(ckey)

# ----------------------------------

# トークン読み込み
with open(token_path, 'r') as f:
    access_tokens = f.read()
import torch
from diffusers import StableDiffusionPipeline
from diffusers.pipelines.stable_diffusion.pipeline_stable_diffusion_img2img import StableDiffusionImg2ImgPipeline, preprocess
from diffusers.pipelines.stable_diffusion.pipeline_stable_diffusion_inpaint import StableDiffusionInpaintPipeline, preprocess_image, preprocess_mask
from PIL import Image
import traceback

# ----------------------------------

# モデルロード
torch_device = "cuda"


def init_text2img_pipe():
    try:
        global text2img_pipes
        # print(f"T: {text2img_pipes[True]}, F: {text2img_pipes[False]}")
        if text2img_pipes[True] and text2img_pipes[False]:
          return
    except Exception as e:
        print(e)
        print(traceback.format_exc())
        text2img_pipes = {True: None, False: None}

    print("■■■■■■■■■■■■■■■")
    print("■　text2image用パイプライン生成開始")
    print("■■■■■■■■■■■■■■■")

    conv2d_init = torch.nn.Conv2d.__init__
    convtranspose2d_init = torch.nn.ConvTranspose2d.__init__

    def init_conv2d_circular(self, *args, **kwargs):
        return conv2d_init(self, *args, **kwargs, padding_mode='circular')

    def init_conv2d_zeros(self, *args, **kwargs):
        return conv2d_init(self, *args, **kwargs, padding_mode='zeros')

    def init_convtranspose2d_circular(self, *args, **kwargs):
        return convtranspose2d_init(self, *args, **kwargs, padding_mode='circular')

    def init_convtranspose2d_zeros(self, *args, **kwargs):
        return convtranspose2d_init(self, *args, **kwargs, padding_mode='zeros')

    try:
        # シームレス用と通常用のパイプラインを生成する
        for b in [True, False]:
            if b:
              torch.nn.Conv2d.__init__ = init_conv2d_circular
              torch.nn.ConvTranspose2d.__init__ = init_convtranspose2d_circular
            else:
              torch.nn.Conv2d.__init__ = init_conv2d_zeros
              torch.nn.ConvTranspose2d.__init__ = init_convtranspose2d_zeros

            # パイプライン生成
            if trinart:
                text2img_pipes[b] = StableDiffusionPipeline.from_pretrained(difussion_model, use_auth_token=access_tokens)  
            else:
                text2img_pipes[b] = StableDiffusionPipeline.from_pretrained(difussion_model, revision="fp16", torch_dtype=torch.float16, use_auth_token=access_tokens)  
            if b:
                print("■■■■■■■■■■■■■■■")
                print("■　text2image（シームレス）用パイプライン生成成功")
                print("■■■■■■■■■■■■■■■")
            else:
                print("■■■■■■■■■■■■■■■")
                print("■　text2image用パイプライン生成成功")
                print("■■■■■■■■■■■■■■■")
    except Exception as e:
        print(e)
        print(traceback.format_exc())

        print("■■■■■■■■■■■■■■■")
        print("■　** ERROR **")
        print("■　学習モデルへのアクセスもしくはセットアップに失敗しました。")
        print("■　　①-B を参考にしながら、学習モデルへのアクセスを許可してください。")
        print("■　　重複して実行すると、学習モデルのセットアップ処理に失敗する場合があります。")
        print("■　　失敗した場合、「ヘッダ＞ランタイム＞セッションの管理」からセッションを一度削除してください。")
        print("■■■■■■■■■■■■■■■")
        raise IpyExit

def init_img2img_pipe():
    try:
        global img2img_pipe
        # print(f"img2img_pipe: {img2img_pipe}")
        if img2img_pipe:
            return
    except Exception as e:
        print(e)
        print(traceback.format_exc())
        img2img_pipe = None

    print("■■■■■■■■■■■■■■■")
    print("■　img2img用パイプライン生成開始")
    print("■■■■■■■■■■■■■■■")

    try:
        # img2img用パイプライン生成
        if trinart:
            img2img_pipe = StableDiffusionImg2ImgPipeline.from_pretrained(difussion_model, use_auth_token=access_tokens)  
        else:
            img2img_pipe = StableDiffusionImg2ImgPipeline.from_pretrained(difussion_model, revision="fp16", torch_dtype=torch.float16, use_auth_token=access_tokens)  

        print("■■■■■■■■■■■■■■■")
        print("■　img2img用パイプライン生成成功")
        print("■■■■■■■■■■■■■■■")
    except Exception as e:
        print(e)
        print(traceback.format_exc())

        print("■■■■■■■■■■■■■■■")
        print("■　** ERROR **")
        print("■　img2img用パイプラインのセットアップに失敗しました。")
        print("■　　重複して実行すると、パイプラインのセットアップ処理に失敗する場合があります。")
        print("■■■■■■■■■■■■■■■")
        raise IpyExit

    return img2img_pipe

def init_inpainting_pipe():
    try:
        global inpainting_pipe
        # print(f"inpainting_pipe: {inpainting_pipe}")
        if inpainting_pipe:
            return
    except Exception as e:
        print(e)
        print(traceback.format_exc())
        inpainting_pipe = None

    print("■■■■■■■■■■■■■■■")
    print("■　inpainting用パイプライン生成開始")
    print("■■■■■■■■■■■■■■■")

    try:
        # inpainting用パイプライン生成
        if trinart:
            inpainting_pipe = StableDiffusionInpaintPipeline.from_pretrained(difussion_model, use_auth_token=access_tokens)  
        else:
            inpainting_pipe = StableDiffusionInpaintPipeline.from_pretrained(difussion_model, revision="fp16", torch_dtype=torch.float16, use_auth_token=access_tokens)  

        print("■■■■■■■■■■■■■■■")
        print("■　inpainting用パイプライン生成成功")
        print("■■■■■■■■■■■■■■■")
    except Exception as e:
        print(e)
        print(traceback.format_exc())

        print("■■■■■■■■■■■■■■■")
        print("■　** ERROR **")
        print("■　inpainting用パイプラインのセットアップに失敗しました。")
        print("■　　重複して実行すると、パイプラインのセットアップ処理に失敗する場合があります。")
        print("■■■■■■■■■■■■■■■")
        raise IpyExit

    return inpainting_pipe

# ----------------------------------

import requests
import json

def translate_jp2en(japanese_text: str):
    if not japanese_text:
      # テキスト入力がない場合、スルー
      return ""

    # 値がないメッセージを翻訳
    params = (
        ("text", japanese_text),
        ("source", "ja"),
        ("target", "en"),
    )

    # GASを叩く
    # https://qiita.com/satto_sann/items/be4177360a0bc3691fdf
    response = requests.get(
        "https://script.google.com/macros/s/AKfycbzZtvOvf14TaMdRIYzocRcf3mktzGgXvlFvyczo/exec",
        params=params,
    )

    # 結果を解析
    results = json.loads(response.text)

    if "text" in results:
        return results["text"]
    else:
        print("■■■■■■■■■■■■■■■")
        print("■　** ERROR **")
        print("■　日本語から英語への翻訳に失敗しました。")
        print("■　少し時間をおいて再度試してみてください。")
        print("■■■■■■■■■■■■■■■")
        raise IpyExit

# ----------------------------------

import os
from datetime import datetime
import torch
from torch import autocast
import ipywidgets as widgets
from ipywidgets import IntSlider, FloatSlider, Label, Layout, HBox, VBox, GridBox, RadioButtons, Text
from IPython.display import display
import functools
import random
from PIL import Image
from tqdm import tqdm

def create_prompt(*args, **kwargs):
    japanese_prompt = kwargs.get("japanese_prompt", "")
    world = kwargs.get("world", "")
    image_type = kwargs.get("image_type", "")
    technique = kwargs.get("technique", "")
    supply = kwargs.get("supply", "")
    school = kwargs.get("school", "")
    photo = kwargs.get("photo", "")
    perspective = kwargs.get("perspective", "")
    camera = kwargs.get("camera", "")
    length = kwargs.get("length", "")
    blur = kwargs.get("blur", "")
    tone = kwargs.get("tone", "")
    hue = kwargs.get("hue", "")
    author = kwargs.get("author", "")
    english_prompt = kwargs.get("english_prompt", "")
    seamless = kwargs.get("seamless", False)
    detail = kwargs.get("detail", False)
    face = kwargs.get("face", False)
    upper = kwargs.get("upper", False)
    lower = kwargs.get("lower", False)
    leg = kwargs.get("leg", False)
    illust = kwargs.get("illust", False)
    aspect = kwargs.get("aspect", "1:1")
    width = kwargs.get("width", 320)
    image_nums = kwargs.get("image_nums", 3)
    num_inference_steps = kwargs.get("num_inference_steps", 50)
    guidance_scale = kwargs.get("guidance_scale", 7.5)
    manual_seed = kwargs.get("manual_seed", 0)
    

    # 画像サイズは必須
    aspect_x, aspect_y = aspect.split(":")
    height = int(int(width) / int(aspect_x) * int(aspect_y))
    jp_prompts = []
    en_prompts = []
    detail_prompts = []
    human_prompts = []

    # 日本語Promptは全角空白で区切って翻訳して設定
    jp_prompts.extend(japanese_prompt.split("　"))

    # 英語Promptはそのまま設定
    en_prompts.append(english_prompt)

    for target, suffix in [(world, ""), (image_type, ""), (author, "が作った"), (supply, "で描いた"), (technique, ""), 
                           (perspective, ""), (photo, ""), (camera, ""), (length, ""), (blur, ""), (tone, ""), (hue, "の色味が強い"), (school, "")]:
        if target and len(target.split(':')):
            # プロンプトが指定されているものだけ追加
            jp_prompts.append(f"{target.split(':')[0]}{suffix}")

    if face or upper or lower:
        human_prompts.extend(["human"])

    if face:
        human_prompts.extend(["face", "head", "hair", "forehead", "temple", "face", "eye", "eyebrow", "eyelash", "eyelid", "eyeball", "iris", "cornea", "nose", "cheek", "mouth", "lips", "ear", "jaw", "neck"])

    if upper:
        human_prompts.extend(["upper", "hand", "palm", "finger", "arm", "elbow", "wrist", "shoulder", "chest", "finger"])

    if lower:
        human_prompts.extend(["lower", "waist", "back", "leg", "knee", "hip"])

    if leg:
        human_prompts.extend(["ankle", "toe", "heel"])

    if illust:
        human_prompts.extend([ "kawaii", "illust", "emotional", "pixiv"])

    if detail:
        detail_prompts.append("high detail")
        detail_prompts.append("high quarity")

    # 翻訳したのをリストアップ
    prompts = []
    other_promts = []
    prompts.append(", ".join([f"( {translate_jp2en(p)} )" for p in jp_prompts]))
    if en_prompts and len(en_prompts[0]):
      prompts.append("(" + " ".join(en_prompts) + ")")
      other_promts.append("(" + " ".join(en_prompts) + ")")
    if human_prompts and len(human_prompts[0]):
      prompts.append("(" + " ".join(human_prompts) + ")")
      other_promts.append("(" + " ".join(human_prompts) + ")")
    if detail_prompts and len(detail_prompts[0]):
      prompts.append("(" + ", ".join(detail_prompts) + ")")
      other_promts.append("(" + " ".join(detail_prompts) + ")")

    # シードは0の場合はランダムとする
    seed = manual_seed if manual_seed else random.randint(1, 1024)

    return ", ".join(prompts), seed, jp_prompts, other_promts, int(width), height

# ------------------------
def execute(
    prompt,
    jp_prompt,
    seamless,
    seed,
    aspect,
    width,
    height,
    image_cnts,
    n_steps,
    g_scale,
    saving,
    img2img=False,
    inpainting=False,
    init_image_path=None,
    mask_image_path=None,
    strength=0
):
    # 指示日時
    now_str = f"{datetime.now():%Y%m%d_%H%M%S}"

    if width % 64 or height % 64:
        print("■■■■■■■■■■■■■■■")
        print("■　** ERROR **")
        print("■　アスペクト比に基づいた画像サイズの縦横が64で割り切れない値になっています。")
        print("■　アスペクト比もしくは、画像横幅の指定を見直してください。")
        print(
            f"■　アスペクト比： {aspect}, 画像横幅: {width}({width / 64:.3f}), 画像縦幅: {height}({height / 64:.3f})"
        )
        print("■■■■■■■■■■■■■■■")
        raise IpyExit

    # テクスチャフラグで切替
    generator = torch.Generator(torch_device).manual_seed(seed)
    file_paths = []

    print("■■■■■■■■■■■■■■■")
    print("■　** PROMPT **")
    print(f"■　{prompt}")
    print("■　** SEED **")
    print(f"■　{seed}")
    print("■■■■■■■■■■■■■■■")

    # 初期画像
    init_image = None
    mask_image = None
    if img2img:
        im = Image.open(init_image_path)
        im = im.convert('RGB')
        init_image = preprocess(im)
        init_img2img_pipe()
    elif inpainting:
        im = Image.open(init_image_path)
        init_image = im.convert('RGB')

        im = Image.open(mask_image_path)
        mask_image = im.convert('RGB')

        init_inpainting_pipe()
    else:
        init_text2img_pipe()

    output_dir = "/content/outputs"

    if img2img:
        output_dir = f"/content/outputs/{now_str}"

    # 出力用フォルダ生成
    os.makedirs(output_dir, exist_ok=True)

    # 指示文を出力
    with open(
        f"{output_dir}/StableDiffusionGenerator_{now_str}.txt", "w", encoding="UTF-8"
    ) as f:
        f.write(jp_prompt)
        f.write("\n")
        f.write(prompt)
        f.write("\n")
        f.write(f"{seed}")

    # 画像を生成
    with autocast(torch_device):
        for n in range(image_cnts):
            m = 0
            try:
                # 一括で指定すると負荷が高くなるので繰り返す
                if img2img:
                    img2img_pipe.to(torch_device)
                    # NSFWの制限を外す
                    img2img_pipe.safety_checker = lambda images, **kwargs: (images, False)
                    # strength が 0の場合、経過で変化させる
                    sg = strength if strength else (min(1, (1 / image_cnts) * (n + 1)))
                    print(f"strength: {sg}")
                    image = img2img_pipe(
                        prompt=prompt,
                        init_image=init_image,
                        strength=sg,
                        guidance_scale=g_scale,
                        num_inference_steps=n_steps,
                    )["sample"][0]
                elif inpainting:
                    inpainting_pipe.to(torch_device)
                    # NSFWの制限を外す
                    inpainting_pipe.safety_checker = lambda images, **kwargs: (images, False)
                    # strength が 0の場合、経過で変化させる
                    sg = strength if strength else (min(1, (1 / image_cnts) * (n + 1)))
                    print(f"strength: {sg}")
                    image = inpainting_pipe(
                        prompt=prompt,
                        init_image=init_image,
                        mask_image=mask_image,
                        strength=sg,
                        guidance_scale=g_scale,
                        num_inference_steps=n_steps,
                    )["sample"][0]
                else:
                    # text2img
                    text2img_pipes[seamless].to(torch_device)
                    if saving:
                        # 省エネモード
                        text2img_pipes[seamless].enable_attention_slicing()
                    else:
                        text2img_pipes[seamless].disable_attention_slicing()
                    # NSFWの制限を外す
                    text2img_pipes[seamless].safety_checker = lambda images, **kwargs: (images, False)
                    image = text2img_pipes[seamless](
                        prompt,
                        width=w,
                        height=h,
                        num_inference_steps=n_steps,
                        guidance_scale=g_scale,
                        generator=generator,
                    )["sample"][0]
            except ValueError as e:
                print("■■■■■■■■■■■■■■■")
                print("■　** ERROR **")
                print("■　与えられたパラメーターが正しくなくて、画像生成に失敗しました。")
                print(f"■　エラーメッセージ：{e}")
                print("■　---------------")
                print("■　　メッセージ別対応方法")
                print(
                    "■　　　ValueError: `height` and `width` have to be divisible by 8 ... 画像サイズを任意入力した場合、サイズの縦横両方ともが 64 で割り切れる整数にしてください"
                )
                print("■■■■■■■■■■■■■■■")
                raise IpyExit
            except RuntimeError as e:
                print(e)
                print(traceback.format_exc())
                print("■■■■■■■■■■■■■■■")
                print("■　** ERROR **")
                print("■　GPU上での処理時に、画像生成に失敗しました。")
                print("■　---------------")
                print("■　　メッセージ別対応方法")
                print(
                    "■　　　RuntimeError: CUDA out of memory. ... メモリ不足です。実行キャパシティの各値を小さくしてください。"
                )
                print(
                    "■　　　Sizes of tensors must match except in dimension 1.　... アスペクト比に基づいた画像サイズで推定ができませんでした。"
                )
                print(
                    "■　　　RuntimeError: Given groups=1, weight of size　... RGBA画像で画像が指定されています。アルファチャンネルを削除してください。"
                )
                print("■■■■■■■■■■■■■■■")
                raise IpyExit
            except Exception as e:
                print(e)
                print(traceback.format_exc())
                print("■■■■■■■■■■■■■■■")
                print("■　** ERROR **")
                print("■　予期せぬ例外により画像生成に失敗しました。")
                print("■■■■■■■■■■■■■■■")
                raise IpyExit

            file_path = (
                f"{output_dir}/StableDiffusionGenerator_{now_str}_{(n + 1):05d}.png"
            )
            # 保存
            image.save(file_path)
            # パス表示
            print("-------------------------------------")
            print(f"■ {file_path}")
            file_paths.append(file_path)
            # 画像表示
            display(image)

    # セル終了
    finish_cell(ckey)

    return file_paths

# ------------------------
jp_prompts = []
other_promts = []
seed = p_manual_seed = 0
saving = p_saving = False
os.makedirs("/content/outputs", exist_ok=True)

# セル終了
finish_cell(ckey)
```
