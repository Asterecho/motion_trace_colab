## ②-A　通知音のダウンロード

通知音を[効果音ラボ](https://soundeffect-lab.info/)様よりダウンロードして、セルの実行を開始・完了・失敗した時に音が鳴るようにします。

下の【②-A】のセルを実行してください



```python
#@markdown ■■■■■■■■■■■■■■■■■■

#@markdown 【②-A】　通知音のダウンロード

#@markdown ■■■■■■■■■■■■■■■■■■

from enum import Enum

class SoundType(Enum):
    START = 0
    SUCCESS = 1
    FAIL = 2

# ----------------------------------

! wget --no-check-certificate -c "https://soundeffect-lab.info/sound/anime/mp3/sceneswitch1.mp3"
! wget --no-check-certificate -c "https://soundeffect-lab.info/sound/anime/mp3/incorrect1.mp3"
! wget --no-check-certificate -c "https://soundeffect-lab.info/sound/anime/mp3/switch1.mp3"

from IPython.display import Audio, display
def play_sound(sound_type: SoundType, autoplay=True):
    try:
        if sound_type == SoundType.START:
            file_name = "switch1.mp3"
        elif sound_type == SoundType.SUCCESS:
            file_name = "sceneswitch1.mp3"
        else:
            file_name = "incorrect1.mp3"

        display(Audio(file_name, autoplay=autoplay))
    except:
        print("■■■■■■■■■■■■■■■")
        print("■　効果音が再生できませんでした")
        print("■■■■■■■■■■■■■■■")

# ----------------------------------

exec_dict = {
    '②-A': False,
    '②-B': False,
    '②-C': False,
    '②-D': False,
    '②-E': False,
    '③': False,
}

NVIDIA_VISIBLE_DEVICES=all

class IpyExit(SystemExit):
    def __init__(self):
        play_sound(SoundType.FAIL)
        pass

def check_exec_dict(cell_key: str):
    for k in list(exec_dict.keys()):
        if k == cell_key:
            play_sound(SoundType.START)
            return True

        if not exec_dict[k]:
            print("■■■■■■■■■■■■■■■")
            print("■　** ERROR **")
            print(f"■　セル【{k}】が実行されていない可能性があります。")
            print("■　目次から戻って実行してください。")
            print("■■■■■■■■■■■■■■■")
            raise IpyExit

    play_sound(SoundType.START)
    return True

def finish_cell(cell_key: str):
    play_sound(SoundType.SUCCESS)

    print("■■■■■■■■■■■■■■■")
    print("■　** OK **")
    print(f"■　セル【{cell_key}】の実行に成功しました。")
    print("■■■■■■■■■■■■■■■")

    exec_dict[cell_key] = True

# ----------------------------------

# セル実行可否
ckey = "②-A"
check_exec_dict(ckey)

# ----------------------------------

# セル終了

finish_cell(ckey)

print("")
print("")
print("■ セル実行開始")
play_sound(SoundType.START, autoplay=False)

print("")
print("")
print("■ セル実行失敗")
play_sound(SoundType.FAIL, autoplay=False)

print("")
print("")
print("■ セル実行成功")
play_sound(SoundType.SUCCESS, autoplay=False)
```

## ②-E　Tokenの認証と学習モデルのダウンロード

StableDiffusionでTokenを使った認証を行い、必要な学習モデルをダウンロードします。

下の【②-E】のセルを実行して下さい。

少し時間がかかります。セルの周りにローディングがくるくる回っている間は待ってください。

```python
#@markdown ■■■■■■■■■■■■■■■■■■

#@markdown 【②-E】　Tokenの認証と学習モデルのダウンロード

#@markdown ■■■■■■■■■■■■■■■■■■

# セル実行可否
ckey = "②-E"
check_exec_dict(ckey)

# ----------------------------------

# トークン読み込み
with open(token_path, 'r') as f:
    access_tokens = f.read()

# モデルロード
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import traceback

# ----------------------------------

torch_device = "cuda" if torch.cuda.is_available() else "cpu"
COMP_VIS_VERSION = "CompVis/stable-diffusion-v1-4"

pipes = {}
conv2d_init = torch.nn.Conv2d.__init__
convtranspose2d_init = torch.nn.ConvTranspose2d.__init__

def init_conv2d_circular(self, *args, **kwargs):
    return conv2d_init(self, *args, **kwargs, padding_mode='circular')

def init_conv2d_zeros(self, *args, **kwargs):
    return conv2d_init(self, *args, **kwargs, padding_mode='zeros')

def init_convtranspose2d_circular(self, *args, **kwargs):
    return convtranspose2d_init(self, *args, **kwargs, padding_mode='circular')

def init_convtranspose2d_zeros(self, *args, **kwargs):
    return convtranspose2d_init(self, *args, **kwargs, padding_mode='zeros')

# load pretrain model
try:
    # make sure you're logged in with `huggingface-cli login`
    for b in [False, True]:
        if b:
          torch.nn.Conv2d.__init__ = init_conv2d_circular
          torch.nn.ConvTranspose2d.__init__ = init_convtranspose2d_circular
        else:
          torch.nn.Conv2d.__init__ = init_conv2d_zeros
          torch.nn.ConvTranspose2d.__init__ = init_convtranspose2d_zeros

        # パイプライン生成
        pipes[b] = StableDiffusionPipeline.from_pretrained(COMP_VIS_VERSION, revision="fp16", torch_dtype=torch.float16, use_auth_token=access_tokens)  
except Exception as e:
    print(e)
    print(traceback.format_exc())

    print("■■■■■■■■■■■■■■■")
    print("■　** ERROR **")
    print("■　学習モデルへのアクセスに失敗しました。")
    print("■　①-B を参考にしながら、学習モデルへのアクセスを許可してください。")
    print("■■■■■■■■■■■■■■■")
    raise IpyExit

# ----------------------------------

import requests
import json

def translate_jp2en(japanese_text: str):
    # 値がないメッセージを翻訳
    params = (
        ("text", japanese_text),
        ("source", "ja"),
        ("target", "en"),
    )

    # GASを叩く
    # https://qiita.com/satto_sann/items/be4177360a0bc3691fdf
    response = requests.get(
        "https://script.google.com/macros/s/AKfycbzZtvOvf14TaMdRIYzocRcf3mktzGgXvlFvyczo/exec",
        params=params,
    )

    # 結果を解析
    results = json.loads(response.text)

    if "text" in results:
        return results["text"]
    else:
        print("■■■■■■■■■■■■■■■")
        print("■　** ERROR **")
        print("■　日本語から英語への翻訳に失敗しました。")
        print("■　少し時間をおいて再度試してみてください。")
        print("■■■■■■■■■■■■■■■")
        raise IpyExit

# ----------------------------------

# セル終了
finish_cell(ckey)
```

# ③　Stable Diffusion 実行


下の【③】を実行すると、`image_nums` に指定された枚数の画像を生成します。

生成したい画像イメージ（4Kとかアニメ風）を指定するとそのイメージに合った画像を作成します。

タッチなどのプルダウンは右端の小さな▼で選択肢が選べます。（任意入力OKの項目は自由に日本語指定可能です）

画像生成枚数などのスライダーは右端に小さく現在の数値が出ています。

```python
#@markdown ■■■■■■■■■■■■■■■■■■

#@markdown 【③】　画像の生成

#@markdown ■■■■■■■■■■■■■■■■■■

#@markdown ### 画像のイメージ

#@markdown 日本語文章や空白区切りの単語など、自由に入力してください

japanese_prompt = "\u53EF\u611B\u304F\u7B11\u304F\u30DC\u30D6\u30AB\u30C3\u30C8\u306E\u5C11\u5973" #@param {type:"string"}

#@markdown ### 種類（選択・任意入力OK）
image_type = "\u5C11\u5973\u6F2B\u753B" #@param ["", "風景画", "静物画", "肖像画", "博物画", "宗教画", "歴史画", "山水画", "風俗画", "美人画", "花鳥画", "写真", "ポスター", "カートゥーン", "アメリカンコミック", "少女漫画", "少年漫画", "アニメ", "古い映画", "カリカチュア", "シルエット", "ポートレート"] {allow-input: true}

#@markdown ### 技法（選択・任意入力OK）
technique = "" #@param ["", "日本画", "洋画", "油彩", "水墨画", "版画", "アクリル画", "アクリルガッシュ", "ペン画", "コラージュ", "セル画", "切り絵", "砂絵", "モザイク", "鉛筆画", "水彩画", "ちぎり絵"] {allow-input: true}

#@markdown ### 画派（選択・任意入力OK）
school = "" #@param ["", "古典主義", "新古典主義", "ロマン主義", "写実主義", "印象派", "象徴主義", "アール・ヌーヴォー", "フォービズム", "キュービズム", "ダダイスム", "ポップアート", "オプアート", "ミニマルアート", "コンセプチュアル・アート", "ボタニカルアート", "大和絵", "萌え絵", "浮世絵", "源氏絵"] {allow-input: true}

#@markdown ### アーティスティック（選択・任意入力OK）
artistic = "" #@param ["", "ポスタリゼーション", "カットアウト", "こする", "スポンジ", "ドライブラシ", "ネオン光彩", "パレットナイフ", "フレスコ", "ラップ", "色鉛筆", "粗いパステル画", "粗描き", "塗料", "粒状フィルム", "シズル感", "ソフトフォーカス", "ノイズ"] {allow-input: true}

#@markdown ### スケッチ（選択・任意入力OK）
sketch = "" #@param ["", "ウォーターペーパー", "ぎざぎざのエッジ", "グラフィックペン", "クレヨンのコンテ画", "クロム", "スタンプ", "チョーク", "木炭画", "ちりめんじわ", "ノート用紙", "ブラスター", "ハーフトーンパターン", "浅浮彫り", "アンティーク", "手ぶれ"] {allow-input: true}

#@markdown ### テクスチャ（選択・任意入力OK）
texture = "" #@param ["", "クラッキング", "ステンドグラス", "テクスチャライザー", "パッチワーク", "モザイクタイル", "粒状"] {allow-input: true}

#@markdown ### ブラシストローク（選択・任意入力OK）
stroke = "" #@param ["", "インク画", "エッジの強調", "スプレーストローク", "斜めストローク", "墨絵", "網目"] {allow-input: true}

#@markdown ### 視点（選択・任意入力OK）
perspective = "\u30D0\u30B9\u30C8\u30A2\u30C3\u30D7" #@param ["", "煽り", "俯瞰", "近景", "遠景", "広角", "望遠", "魚眼", "バストアップ", "顔写真", "全身", "ポートレート"] {allow-input: true}

#@markdown ### 色調（選択・任意入力OK）
tone = "" #@param ["", "ビビッドトーン", "ブライトトーン", "ストロングトーン", "ディープトーン", "ライトトーン", "ソフトトーン", "ダルトーン", "ダークトーン", "ペールトーン", "ライトグレイッシュトーン", "グレイッシュトーン", "ダークグレイッシュトーン", "ベリーペールトーン", "ペールグレイッシュトーン", "ミドルグレイッシュトーン", "ベリーグレイッシュトーン", "ベリーダークトーン"] {allow-input: true}

#@markdown ### 色相（選択・任意入力OK）
hue = "" #@param ["", "赤みがかった", "青みがかった", "緑みがかった", "黄みがかった", "紫みがかった", "橙みがかった", "黒みがかった", "灰みがかった", "白みがかった", "モノクロ", "セピア"] {allow-input: true}

#@markdown ### 特定画家の画風（選択・任意入力OK）
painting = "" #@param ["", "ピカソ", "ゴッホ", "モネ", "ミュシャ", "葛飾北斎", "新海誠", "宮崎駿夫"] {allow-input: true}

#@markdown ### 追加Prompt（翻訳なし）

english_prompt = "" #@param {type:"string"}

#@markdown ### 画像の特徴

#@markdown シームレステクスチャ画像であるか否か

texture = False #@param {type:"boolean"}

#@markdown タイリング（絵柄が敷き詰められている）画像であるか否か

tiling = False #@param {type:"boolean"}

#@markdown 人物の顔パーツを指定するか

face = True #@param {type:"boolean"}

#@markdown 人物の上半身パーツを指定するか

upper = True #@param {type:"boolean"}

#@markdown 人物の下半身パーツを指定するか

lower = False #@param {type:"boolean"}

#@markdown イラスト風人物（萌え系）

illust = False #@param {type:"boolean"}

#@markdown ### 画像サイズ

#@markdown アスペクト比

aspect = "2:1"  #@param ['1:1', '2:1', '3:2', '4:3', '16:9', '1:2', '2:3', '3:4', '9:16']

#@markdown 画像横幅（縦幅はアスペクト比を加味して算出します。任意入力OKですが、縦横いずれも64の倍数になるようにしてください。）

width = "768" #@param ["320", "768", "1280", "1920", "2560", "3840"] {allow-input: true}

#@markdown ### 画像生成枚数

#@markdown （値が大きいほど沢山の画像が生成されます）

image_nums = 6 #@param {type:"slider", min:1, max:20}

#@markdown ### 推定パラメーター

#@markdown （値が大きいほどイメージに合った画像が出てくる可能性が高くなりますが、その分負荷も高くなります）

num_inference_steps = 40 #@param {type:"slider", min:1, max:200}

guidance_scale = 7.5 #@param {type:"slider", min:0.1, max:20, step: 0.1}

# セル実行可否
ckey = "③"
check_exec_dict(ckey)

# ----------------------------------

import os
from datetime import datetime
import torch
from torch import autocast
from IPython.display import display

def create_prompt():
    # 画像サイズは必須
    aspect_x, aspect_y = aspect.split(":")
    height = int(int(width) / int(aspect_x) * int(aspect_y))
    args = []
    prompts = []

    for target in [image_type, technique, perspective, tone, hue, painting]:
        if target:
            # プロンプトが指定されている場合、翻訳して追加
            prompts.append(translate_jp2en(target))

    for target in [school, artistic, sketch, texture, stroke, painting]:
        if target:
            # プロンプトが指定されている場合、翻訳して追加
            prompts.append(translate_jp2en(f"{target}のような"))

    if texture:
        prompts.append("texture")

    if tiling:
        prompts.append("tiling")

    if face or upper or lower:
        prompts.append(" ".join(["human"]))

    if face:
        prompts.append(" ".join(["face", "head", "hair", "forehead", "temple", "face", "eye", "eyebrow", "eyelash", "eyelid", "eyeball", "iris", "cornea", "nose", "cheek", "mouth", "lips", "ear", "jaw", "neck"]))

    if upper:
        prompts.append(" ".join(["upper", "hand", "palm", "finger", "arm", "elbow", "wrist", "shoulder", "chest", "finger"]))

    if lower:
        prompts.append(" ".join(["lower", "waist", "back", "leg", "knee", "ankle", "toe", "heel"]))

    if illust:
        prompts.append(" ".join([ "kawaii", "pixiv"]))

    prompts.append(translate_jp2en(japanese_prompt))

    # 英語Promptはそのまま設定
    prompts.append(english_prompt)

    return ", ".join(prompts), int(width), height

# 指示日時
now_str = f"{datetime.now():%Y%m%d_%H%M%S}"

# 出力用フォルダ生成
os.makedirs(f"outputs", exist_ok=True)

# prompt生成
prompt, w, h = create_prompt()

if w % 64 or h % 64:
    print("■■■■■■■■■■■■■■■")
    print("■　** ERROR **")
    print("■　アスペクト比に基づいた画像サイズの縦横が64で割り切れない値になっています。")
    print("■　アスペクト比もしくは、画像横幅の指定を見直してください。")
    print(f"■　アスペクト比： {aspect}, 画像横幅: {w}({w / 64:.3f}), 画像縦幅: {h}({h / 64:.3f})")
    print("■■■■■■■■■■■■■■■")
    raise IpyExit

# 指示文を出力
with open(f"outputs/StableDiffusionGenerator_{now_str}.txt", "w", encoding="UTF-8") as f:
    f.write(japanese_prompt)
    f.write("\n")
    f.write(prompt)

print("■■■■■■■■■■■■■■■")
print("■　** PROMPT **")
print(f"■　{prompt}")
print("■■■■■■■■■■■■■■■")

# テクスチャフラグで切替
pipes[texture].to(torch_device)

# 画像を生成
with autocast(torch_device):
    for n in range(image_nums):
        try:
            # 一括で指定すると負荷が高くなるので繰り返す
            image = pipes[texture](prompt, width=w, height=h, num_inference_steps=num_inference_steps, guidance_scale=guidance_scale)["sample"][0]
        except ValueError as e:
            print("■■■■■■■■■■■■■■■")
            print("■　** ERROR **")
            print("■　与えられたパラメーターが正しくなくて、画像生成に失敗しました。")
            print(f"■　エラーメッセージ：{e}")
            print("■　---------------")
            print("■　　メッセージ別対応方法")
            print("■　　　ValueError: `height` and `width` have to be divisible by 8 ... 画像サイズを任意入力した場合、サイズの縦横両方ともが 64 で割り切れる整数にしてください")
            print("■■■■■■■■■■■■■■■")
            raise IpyExit
        except RuntimeError as e:
            print(e)
            print(traceback.format_exc())
            print("■■■■■■■■■■■■■■■")
            print("■　** ERROR **")
            print("■　GPU上での処理時に、画像生成に失敗しました。")
            print("■　---------------")
            print("■　　メッセージ別対応方法")
            print("■　　　RuntimeError: CUDA out of memory. ... メモリ不足です。実行キャパシティの各値を小さくしてください。")
            print("■　　　Sizes of tensors must match except in dimension 1.　... アスペクト比に基づいた画像サイズで推定ができませんでした。")
            print("■■■■■■■■■■■■■■■")
            raise IpyExit
        except Exception as e:
            print(e)
            print(traceback.format_exc())
            print("■■■■■■■■■■■■■■■")
            print("■　** ERROR **")
            print("■　予期せぬ例外により画像生成に失敗しました。")
            print("■■■■■■■■■■■■■■■")
            raise IpyExit

        # 保存
        image.save(f"outputs/StableDiffusionGenerator_{now_str}_{n}.png")
        # 表示
        display(image)

# ----------------------------------

# セル終了
finish_cell(ckey)
```
